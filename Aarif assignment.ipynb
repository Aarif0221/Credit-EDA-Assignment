```python
#To find null values:
def null_percentage(df):
    output = round(df.isnull().sum()/len(df.index)*100,2)
    return output
```

```python
def dropnullcol (appl_data1 ,percentage):
    colnames=(appl_data1.isnull().sum()/appl_data1.shape[0]*100)
    colnames=list(colnames[colnames.values>=percentage].index)
    appl_data1.drop(labels=colnames,axis=1,inplace=True)
    print("The no. of columns dropped:",len(colnames))
    return appl_data1
```

```python
#Use a mix of univariate and bivariate analysis etc.
#univariate analysis for categorical variables:
def plot_uni_cat(col):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    plt.title('Defaulter')
    sns.countplot(x=target_11[col])
    plt.subplot(1,2,2)
    plt.title('Non_Defaulter')
    sns.countplot(x=target_12[col])
```

```python
#for numeric columns:
def plot_uni_num(col):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    plt.title('Defaulter')
    sns.distplot(x=target_11[col],hist=False)
    plt.subplot(1,2,2)
    plt.title('Non_Defaulter')
    sns.distplot(x=target_12[col],hist=False)
```
